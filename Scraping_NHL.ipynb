#Dataframe con los datos de las estadsticas

```
soup.select('tr > td') #podemos visualizar las filas de los datos que nos interesa
```

```
rows=soup.find_all('tr') # nos quedamos con todo el bloque que contiene las filas td de inters
```

```
list_rows= []

for row in rows:
  row_td=row.find_all('td')
  str_cells=str(row_td)
  clean_text=BeautifulSoup(str_cells, "lxml").get_text()
  list_rows.append(clean_text)
```

```
import pandas as pd
data=pd.DataFrame(list_rows)
data.head(5)
```

```
data1=data[0].str.split(pat=", ", expand=True)
```

```
data1.head()
```

```
data1 = data1.astype(str).replace({"\[":"", "\]":""}, regex=True)
```

```
data_final=pd.DataFrame(data=data1)
```

```
data_final=data_final.drop([0,1], axis=0)
```

```
data_final.head()
```

```
data_final.reset_index(drop=True, inplace=True)
```

```
data_final.columns = ['Team',	'Age',	'Pos',	'GP',	'G',	'A',	'P', 'PIM',	'+/-',	'TOI',	'ES',	'PP',	'SH',	'ESG',	'PPG',	'SHG',	'GWG',	'OTG',
                      'ESA',	'PPA'	,'SHA',	'GWA'	,'OTA',	'ESP'	,'PPP',	'SHP'	,'GWP',	'OTP',	'PPP%',	'G/60',	'A/60',	'P/60'	,'ESG/60',	'ESA/60',
                      'ESP/60',	'PPG/60',	'PPA/60',	'PPP/60',	'G/GP',	'A/GP',	'P/GP',	'SHOTS',	'SH%',	'HITS',	'BS',	'FOW',	'FOL',	'FO%']
```

```
soup.select('tr > th')
```

```
rows=soup.find_all('tr') 
```

```
list_rows= []

for row in rows:
  row_td=row.find_all('th')
  str_cells=str(row_td)
  clean_text=BeautifulSoup(str_cells, "lxml").get_text()
  list_rows.append(clean_text)
```

```
jugadores=pd.DataFrame(list_rows)
jugadores.head(5)
```

```
jugadores=jugadores[0].str.split(pat=", ", expand=True)
```

```
jugadores.head()
```

```
jugadores = jugadores.drop([0,1], axis = 0)
```

```
jugadores.columns = ['Nombre']
jugadores.head()
```
